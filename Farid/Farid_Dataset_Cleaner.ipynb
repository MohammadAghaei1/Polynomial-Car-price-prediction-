```python
import pandas as pd
import numpy as np
```

```python
with open('/content/Polynomial-Car-price-prediction-/Dataset/Car data.csv', 'r') as f:
    lines = f.readlines()

data = [line.strip().split(',') for line in lines]

# Assuming the first row is the header
header = [col.strip().replace('�"', '').replace('"', '') for col in data[0]]
data = data[1:]

df = pd.DataFrame(data, columns=header)
display(df.head())
display(df.info())
```

```python
df.rename(columns={'﻿Car_Name': 'Car_Name'}, inplace=True)
df['Car_Name'] = df['Car_Name'].str.replace('"', '')
df['Owner'] = df['Owner'].str.replace('"', '')
display(df.head())
```

```python
df.to_csv('car_data_cleaned.csv', index=False)
```
